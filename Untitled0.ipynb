```python

import os
import shutil
import requests
import re
import difflib
from graphviz import Digraph
from IPython.display import Image, display
from cairosvg import svg2png

# ----------------------------
# Step 0: Setup cache folder
# ----------------------------
CACHE_FOLDER = "icon_cache"
if os.path.exists(CACHE_FOLDER):
    shutil.rmtree(CACHE_FOLDER)
os.makedirs(CACHE_FOLDER)

BASE_URL = "https://storage.googleapis.com/eraser-public-assets/canvas-icons/"

# ----------------------------
# Step 1: Fetch available icons from the server
# ----------------------------
def fetch_available_icons():
    """Fetch list of available icons from the server"""
    try:
        index_url = BASE_URL + "index.json"
        response = requests.get(index_url, timeout=10)
        if response.status_code == 200:
            icons = response.json()
            return {icon['name'].lower().replace('.svg', ''): BASE_URL + icon['name']
                   for icon in icons if icon['name'].endswith('.svg')}
        else:
            # print(f"âš ï¸ Could not fetch icon index from {index_url}")
            return {}
    except Exception as e:
        # print(f"âš ï¸ Error fetching icon index: {e}")
        return {}

# ----------------------------
# Step 2: Build dynamic icon map (AWS + non-AWS)
# ----------------------------
def build_dynamic_icon_map(user_prompt):
    """Build icon map based on entities found in user prompt"""
    entities = []
    chains = [chain.strip() for chain in user_prompt.split(",")]
    for chain in chains:
        parts = [p.strip() for p in re.split(r"->|=>", chain)]
        for part in parts:
            if part and part not in entities:
                entities.append(part)

    # print(f"ðŸ” Found entities in prompt: {entities}")

    available_icons = fetch_available_icons()
    icon_map = {}

    for entity in entities:
        entity_lower = entity.lower()
        possible_names = []

        # AWS-specific icons
        aws_services = {
            'quicksight': 'aws-quicksight',
            'sagemaker': 'aws-sagemaker',
            's3': 'aws-simple-storage-service',
            'S3-Bucket': 'aws-simple-storage-service',
            'glue': 'aws-glue',
            'lambda': 'aws-lambda',
            'ec2': 'aws-ec2',
            'rds': 'aws-rds',
            'dynamodb': 'aws-dynamodb',
            'api gateway': 'aws-api-gateway',
            'cloudfront': 'aws-cloudfront',
            'route53': 'aws-route53',
            'vpc': 'aws-vpc',
            'iam': 'aws-iam',
            'cloudwatch': 'aws-cloudwatch',
            'elb': 'aws-elb',
            'alb': 'aws-alb',
            'efs': 'aws-efs',
            'emr': 'aws-emr',
            'redshift': 'aws-redshift',

            'kinesis': 'aws-kinesis'
        }
        if entity_lower in aws_services:
            possible_names.append(aws_services[entity_lower])

        # Generic / non-AWS
        possible_names.extend([
            entity_lower,
            entity_lower.replace(" ", "-"),
            entity_lower.replace(" ", "_"),
            entity_lower.replace(" ", ""),
            entity_lower.replace("server", ""),
            entity_lower.replace("database", ""),
            entity_lower.replace("app", "")
        ])

        # Special cases
        if entity_lower == 'user':
            possible_names.extend(['user', 'person', 'client', 'customer', 'end-user'])
        elif entity_lower == 'aws':
            possible_names.extend(['aws', 'amazon-web-services', 'amazon'])

        # print(f"ðŸ” Trying icon names for '{entity}': {possible_names[:5]}...")

        found_icon = None

        # Exact match in available icons
        if available_icons:
            for name in possible_names:
                if name in available_icons:
                    found_icon = available_icons[name]
                    # print(f"âœ… Found in index: {name}")
                    break

            # Fuzzy match
            if not found_icon:
                matches = difflib.get_close_matches(entity_lower, available_icons.keys(), n=1, cutoff=0.6)
                if matches:
                    found_icon = available_icons[matches[0]]
                    # print(f"ðŸ” Fuzzy match for '{entity}': {matches[0]}")

        # Fallback: test direct URL
        if not found_icon:
            # print(f"ðŸ“¡ Testing direct URLs for '{entity}'...")
            for name in possible_names:
                potential_url = f"{BASE_URL}{name}.svg"
                try:
                    response = requests.head(potential_url, timeout=5)
                    if response.status_code == 200:
                        found_icon = potential_url
                        # print(f"âœ… Found icon by direct URL: {name}.svg")
                        break
                except requests.RequestException:
                    continue

        if found_icon:
            icon_map[entity_lower] = found_icon
            # print(f"ðŸ“Œ Mapped '{entity}' -> {os.path.basename(found_icon)}")
        else:
            # print(f"âŒ No icon found for '{entity}' after trying {len(possible_names)} variations")
            pass

    return icon_map

# ----------------------------
# Step 3: Download icon dynamically
# ----------------------------
def download_icon(name, icon_map):
    """Download and convert icon to PNG"""
    if not icon_map:
        return None

    name_lower = name.lower()
    if name_lower in icon_map:
        icon_url = icon_map[name_lower]
    else:
        match = difflib.get_close_matches(name_lower, icon_map.keys(), n=1, cutoff=0.5)
        if not match:
            # print(f"âš ï¸ No icon mapping found for '{name}'")
            return None
        icon_url = icon_map[match[0]]
        # print(f"ðŸ” Using fuzzy match: '{name}' -> '{match[0]}'")

    filename = os.path.basename(icon_url)
    svg_path = os.path.join(CACHE_FOLDER, filename)
    png_path = svg_path.replace(".svg", ".png")

    if not os.path.exists(png_path):
        try:
            # print(f"ðŸ“¥ Downloading: {icon_url}")
            r = requests.get(icon_url, timeout=10)
            r.raise_for_status()
            with open(svg_path, "wb") as f:
                f.write(r.content)

            svg2png(url=svg_path, write_to=png_path, output_width=64, output_height=64)
            # print(f"âœ… Downloaded & converted: {filename} for '{name}'")
            if os.path.exists(svg_path):
                os.remove(svg_path)
        except Exception:
            return None

    return png_path

# ----------------------------
# Step 4: Parse user prompt to JSON
# ----------------------------
def parse_user_prompt_to_json(user_prompt, icon_map, title="Dynamic Architecture"):
    entities = {}
    relationships = []
    chains = [chain.strip() for chain in user_prompt.split(",")]

    for chain in chains:
        parts = [p.strip() for p in re.split(r"->|=>", chain)]
        for i, part in enumerate(parts):
            if not part:
                continue
            ent_id = part.lower().replace(" ", "_").replace("-", "_")
            if ent_id not in entities:
                icon_path = download_icon(part, icon_map)
                entities[ent_id] = {
                    "id": ent_id,
                    "label": part,
                    "icon": icon_path
                }
            if i > 0 and parts[i-1].strip():
                src_id = parts[i-1].strip().lower().replace(" ", "_").replace("-", "_")
                tgt_id = ent_id
                rel = {"source": src_id, "target": tgt_id, "label": ""}
                if rel not in relationships:
                    relationships.append(rel)

    return {
        "title": title,
        "entities": list(entities.values()),
        "relationships": relationships
    }

# ----------------------------
# Step 5: Recursive entity adder
# ----------------------------
def add_entity(dot, ent):
    if ent.get("cluster"):
        with dot.subgraph(name=f"cluster_{ent['id']}") as c:
            c.attr(label=ent["label"], fontsize="12", style="rounded,dashed", color="gray")
            for child in ent.get("children", []):
                add_entity(c, child)
    else:
        icon_path = ent.get("icon")
        if icon_path and os.path.exists(icon_path):
            html_label = f'''<


                {ent["label"]}

            >'''
            dot.node(ent["id"], label=html_label, shape="none", margin="0.1")
        else:
            dot.node(
                ent["id"],
                label=ent["label"],
                shape="box",
                style="rounded,filled",
                fillcolor="lightblue",
                color="black",
                fontname="Helvetica",
                fontsize="10"
            )

# ----------------------------
# Step 6: Generate diagram
# ----------------------------
def generate_architecture_diagram_from_prompt(user_prompt, title=None):
    if title is None:
        title = "Architecture Diagram"
    try:
        # print("ðŸ”§ Building dynamic icon map...")
        icon_map = build_dynamic_icon_map(user_prompt)
        # print("ðŸ“ Parsing user prompt...")
        json_data = parse_user_prompt_to_json(user_prompt, icon_map, title)
        if not json_data.get("entities"):
            # print("âš ï¸ No entities found in the prompt")
            return

        dot = Digraph("Architecture", format="png")
        dot.attr(rankdir="LR", splines="ortho", bgcolor="white", pad="0.5", nodesep="1.0", ranksep="1.5")
        dot.attr("node", fontname="Helvetica", fontsize="10")
        dot.attr("edge", fontname="Helvetica", fontsize="9", color="gray40")

        if json_data.get("title"):
            dot.attr(label=json_data["title"], labelloc="t", fontsize="16", fontname="Helvetica-Bold")

        for ent in json_data.get("entities", []):
            add_entity(dot, ent)

        for rel in json_data.get("relationships", []):
            attrs = {"fontsize": "9", "labeldistance": "2.0", "labelangle": "0", "arrowsize": "0.8"}
            dot.edge(rel["source"], rel["target"], label=rel.get("label", ""), **attrs)

        # print("ðŸŽ¨ Rendering diagram...")
        png_bytes = dot.pipe(format="png")
        display(Image(png_bytes))
        output_file = "architecture_diagram.png"
        with open(output_file, "wb") as f:
            f.write(png_bytes)
        # print(f"ðŸ’¾ Diagram saved as: {output_file}")

    except Exception:
        raise
    finally:
        if os.path.exists(CACHE_FOLDER):
            shutil.rmtree(CACHE_FOLDER)
            os.makedirs(CACHE_FOLDER)
            # print("ðŸ—‘ï¸ Icon cache cleared.")

# ----------------------------
# Step 7: Example usage
# ----------------------------
if __name__ == "__main__":
    # print("=" * 60)
    # print("Dynamic AWS + Non-AWS Icon Fetching Example")
    # print("=" * 60)
    # user_prompt = "User -> Redis -> API Gateway -> Redis -> API Gateway -> S3 -> SageMaker -> QuickSight -> azure-cost-budgets -> User"
    # user_prompt = "DataSet -> S3 -> Glue DataBrew -> d- S3, Glue DataBrew -> aws-glue-elastic-views -> aws-athena -> SageMaker -> CloudWatch, SageMaker -> QuickSight"
    user_prompt = "User -> S3 -> AWS Glue -> D-S3 -> Athena -> SageMaker -> CloudWatch, Quicksight"
    generate_architecture_diagram_from_prompt(user_prompt, "Generic Pipeline")
```

```python
# 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

# ==========================
# Install deps (Colab)
# ==========================


import os
import re
import shutil
import requests
import difflib
from groq import Groq
from graphviz import Digraph
from IPython.display import Image, display
from cairosvg import svg2png

# ==========================
# Groq API config
# ==========================
os.environ["GROQ_API_KEY"] = os.environ.get("GROQ_API_KEY", "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")
groq_client = Groq(api_key=os.environ["GROQ_API_KEY"])

# ==========================
# System prompt for LLaMA 90B
# ==========================
SYSTEM_PROMPT = """
You are an expert in converting natural language architecture or data flow descriptions into a **precise, standardized flow string**.

Rules:

1. **Sequential Flow**
   - Use "->" to indicate the normal sequential flow from one component to another.
   - Example: `A -> B -> C`

2. **Parallel / Branching Flow**
   - If a component sends data to multiple destinations, **repeat the source component before each branch**.
   - Example: `A -> B, A -> C, A -> D`
   - If after the branches the flow continues, extend each branch separately:
     - Example: `A -> B -> X, A -> C -> X, A -> D -> X`

3. **Loops / Back-edges**
   - Preserve any loops or backward flows exactly as described.
   - Example: `A -> B -> C -> B`

4. **Entity Names**
   - Keep entity names exactly as they appear in the description.
   - Remove the prefix "Amazon" or "AWS" from services (e.g., "Amazon S3" â†’ "S3", "AWS Glue" â†’ "Glue").
   - Do NOT append descriptive text to service names (e.g., donâ€™t add "bucket", "function", etc.).
   - Make sure "User" is always the first entity if mentioned.

5. **Output Format**
   - Output **only** the flow string.
   - No explanations, no JSON, no markdown.
   - Multiple branches must be separated by commas as separate flows.

6. **Examples**
   - Description: "A user uploads data into S3, then AWS Glue processes the data and stores results into another S3 bucket called D-S3. The processed data in D-S3 is queried using Athena, and the query results are sent to SageMaker for machine learning. The output of SageMaker is monitored in CloudWatch and also visualized in QuickSight."
     - Output: `User -> S3 -> Glue -> D-S3, Glue -> Athena -> SageMaker -> CloudWatch, SageMaker -> QuickSight`

   - Description: "User sends request to S3, which triggers Lambda, then both Lambda and S3 send events to SNS, and finally SNS notifies both SQS and Email."
     - Output: `User -> S3 -> Lambda -> SNS, S3 -> SNS -> SQS, S3 -> SNS -> Email`

7. **Important**
   - Do not interpret or invent components.
   - Do not add "start" or "end".
   - Ensure parallel branches always restart with the parent node before continuing.

Your output must be a clean **flow string** ready for diagram generation.
"""


# ==========================
# Natural prompt -> flow string using Groq LLaMA 90B
# ==========================
def natural_to_flow(user_prompt):
    response = groq_client.chat.completions.create(
        messages=[{"role": "user", "content": f"{SYSTEM_PROMPT}\n\nInput: {user_prompt}\nOutput:"}],
        model="gemma2-9b-it"
    )
    # Access content correctly
    return response.choices[0].message.content.strip()


# ==========================
# Icon cache setup
# ==========================
CACHE_FOLDER = "icon_cache"
if os.path.exists(CACHE_FOLDER):
    shutil.rmtree(CACHE_FOLDER)
os.makedirs(CACHE_FOLDER)

BASE_URL = "https://storage.googleapis.com/eraser-public-assets/canvas-icons/"

def fetch_available_icons():
    try:
        index_url = BASE_URL + "index.json"
        r = requests.get(index_url, timeout=10)
        if r.status_code == 200:
            icons = r.json()
            return {icon['name'].lower().replace('.svg',''): BASE_URL + icon['name']
                    for icon in icons if icon['name'].endswith('.svg')}
    except Exception:
        pass
    return {}

# ==========================
# Icon resolution / download
# ==========================
def build_dynamic_icon_map(flow_string):
    entities = []
    segments = [s.strip() for s in re.split(r'\s*->\s*', flow_string) if s.strip()]
    for seg in segments:
        for part in re.split(r'\s*,\s*', seg):
            p = part.strip()
            if p and p not in entities:
                entities.append(p)

    available_icons = fetch_available_icons()
    icon_map = {}
    aws_services = {
        'quicksight': 'aws-quicksight',
        'sagemaker': 'aws-sagemaker',
        's3': 'aws-simple-storage-service',
        'glue': 'aws-glue',
        'databrew': 'aws-glue-databrew',
        'athena': 'aws-athena',
        'cloudwatch': 'aws-cloudwatch',
        'api gateway': 'aws-api-gateway',
        'redis': 'redis',
        'lambda': 'aws-lambda',
        'cloudfront': 'aws-cloudfront',
        'ses': 'aws-ses'
    }

    for entity in entities:
        entity_lower = entity.lower()
        possible = [aws_services.get(entity_lower, None)]
        possible += [
            entity_lower,
            entity_lower.replace(" ", "-"),
            entity_lower.replace(" ", "_"),
            entity_lower.replace(" ", "")
        ]
        if entity_lower == "user":
            possible += ["user", "person", "client", "customer", "end-user"]

        found = None
        for name in possible:
            if name and name in available_icons:
                found = available_icons[name]
                break
        if not found:
            m = difflib.get_close_matches(entity_lower, available_icons.keys(), n=1, cutoff=0.6)
            if m:
                found = available_icons[m[0]]
        if not found:
            for name in possible:
                try:
                    url = f"{BASE_URL}{name}.svg"
                    head = requests.head(url, timeout=4)
                    if head.status_code == 200:
                        found = url
                        break
                except Exception:
                    continue
        if found:
            icon_map[entity_lower] = found
    return icon_map

def download_icon(name, icon_map):
    if not icon_map:
        return None
    name_l = name.lower()
    url = icon_map.get(name_l)
    if not url:
        m = difflib.get_close_matches(name_l, icon_map.keys(), n=1, cutoff=0.5)
        if not m:
            return None
        url = icon_map[m[0]]
    filename = os.path.basename(url)
    svg_path = os.path.join(CACHE_FOLDER, filename)
    png_path = svg_path.replace('.svg', '.png')
    if not os.path.exists(png_path):
        try:
            r = requests.get(url, timeout=10)
            r.raise_for_status()
            with open(svg_path, 'wb') as f:
                f.write(r.content)
            svg2png(url=svg_path, write_to=png_path, output_width=64, output_height=64)
            if os.path.exists(svg_path):
                os.remove(svg_path)
        except Exception:
            return None
    return png_path

# ==========================
# Parse flow -> JSON
# ==========================
def parse_flow_to_json(flow, icon_map, title="Generated Pipeline"):
    flow = re.sub(r'\s+', ' ', flow.strip())
    segments = [seg.strip() for seg in re.split(r'\s*->\s*', flow) if seg.strip()]
    nodes_per_segment = []
    for seg in segments:
        nodes = [n.strip() for n in re.split(r'\s*,\s*', seg) if n.strip()]
        nodes_per_segment.append(nodes)

    entities = {}
    relationships = []
    entity_counter = {}

    for stage_idx, nodes in enumerate(nodes_per_segment):
        for node_label in nodes:
            base_id = node_label.lower().replace(" ", "_").replace("-", "_")
            count = entity_counter.get(base_id, 0) + 1
            entity_counter[base_id] = count
            ent_id = f"{base_id}_{count}"
            icon_path = download_icon(node_label, icon_map)
            entities[ent_id] = {
                "id": ent_id,
                "label": node_label,
                "icon": icon_path,
                "stage": stage_idx
            }

    for i in range(len(nodes_per_segment)-1):
        src_nodes = nodes_per_segment[i]
        tgt_nodes = nodes_per_segment[i+1]
        for s_idx, s in enumerate(src_nodes):
            s_id = f"{s.lower().replace(' ','_').replace('-','_')}_{entity_counter[s.lower().replace(' ','_').replace('-','_')]}"
            for t_idx, t in enumerate(tgt_nodes):
                t_id = f"{t.lower().replace(' ','_').replace('-','_')}_{entity_counter[t.lower().replace(' ','_').replace('-','_')]}"
                relationships.append({"source": s_id, "target": t_id})

    max_stage = max((e["stage"] for e in entities.values()), default=-1)
    stages = []
    for i in range(max_stage+1):
        stage_nodes = [eid for eid, e in entities.items() if e["stage"] == i]
        stages.append(stage_nodes)

    return {"title": title, "entities": list(entities.values()), "relationships": relationships, "stages": stages}

# ==========================
# Graphviz helper
# ==========================
def add_entity(graph, ent):
    icon_path = ent.get("icon")
    if icon_path and os.path.exists(icon_path):
        html_label = f'''<
            <TABLE BORDER="0" CELLBORDER="0" CELLSPACING="0">
                <TR><TD FIXEDSIZE="TRUE" WIDTH="64" HEIGHT="64"><IMG SRC="{icon_path}"/></TD></TR>
                <TR><TD>{ent["label"]}</TD></TR>
            </TABLE>
        >'''
        graph.node(ent["id"], label=html_label, shape="none", margin="0.1")
    else:
        graph.node(ent["id"], label=ent["label"], shape="box", style="rounded,filled",
                   fillcolor="lightblue", color="black", fontname="Helvetica", fontsize="10")

# ==========================
# Generate diagram
# ==========================
# def generate_architecture_diagram_from_prompt(flow_string, title="Generated Pipeline", save_path="architecture_diagram.png"):
#     icon_map = build_dynamic_icon_map(flow_string)
#     json_data = parse_flow_to_json(flow_string, icon_map, title)
#     entities_map = {e["id"]: e for e in json_data["entities"]}

#     dot = Digraph("Architecture", format="png")
#     dot.attr(rankdir="LR", splines="ortho", bgcolor="white", pad="0.5", nodesep="1.0", ranksep="1.25")
#     dot.attr("node", fontname="Helvetica", fontsize="10")
#     dot.attr("edge", fontname="Helvetica", fontsize="9")
#     if json_data.get("title"):
#         dot.attr(label=json_data["title"], labelloc="t", fontsize="16", fontname="Helvetica-Bold")

#     for stage_idx, stage_nodes in enumerate(json_data.get("stages", [])):
#         with dot.subgraph(name=f"stage_{stage_idx}") as s:
#             s.attr(rank="same")
#             for ent_id in stage_nodes:
#                 add_entity(s, entities_map[ent_id])

#     for rel in json_data.get("relationships", []):
#         src = rel["source"]
#         tgt = rel["target"]
#         src_stage = entities_map.get(src, {}).get("stage", 0)
#         tgt_stage = entities_map.get(tgt, {}).get("stage", 0)
#         if src_stage > tgt_stage:
#             dot.edge(src, tgt, color="red", style="dashed", arrowsize="0.9", constraint="false")
#         else:
#             dot.edge(src, tgt, color="gray40", arrowsize="0.8")

#     png_bytes = dot.pipe(format="png")
#     display(Image(png_bytes))
def generate_architecture_diagram_from_prompt(user_prompt, title=None):
    if title is None:
        title = "Architecture Diagram"
    try:
        # print("ðŸ”§ Building dynamic icon map...")
        icon_map = build_dynamic_icon_map(user_prompt)
        # print("ðŸ“ Parsing user prompt...")
        json_data = parse_user_prompt_to_json(user_prompt, icon_map, title)
        if not json_data.get("entities"):
            # print("âš ï¸ No entities found in the prompt")
            return

        dot = Digraph("Architecture", format="png")
        dot.attr(rankdir="LR", splines="ortho", bgcolor="white", pad="0.5", nodesep="1.0", ranksep="1.5")
        dot.attr("node", fontname="Helvetica", fontsize="10")
        dot.attr("edge", fontname="Helvetica", fontsize="9", color="gray40")

        if json_data.get("title"):
            dot.attr(label=json_data["title"], labelloc="t", fontsize="16", fontname="Helvetica-Bold")

        for ent in json_data.get("entities", []):
            add_entity(dot, ent)

        for rel in json_data.get("relationships", []):
            attrs = {"fontsize": "9", "labeldistance": "2.0", "labelangle": "0", "arrowsize": "0.8"}
            dot.edge(rel["source"], rel["target"], label=rel.get("label", ""), **attrs)

        # print("ðŸŽ¨ Rendering diagram...")
        png_bytes = dot.pipe(format="png")
        display(Image(png_bytes))
        output_file = "architecture_diagram.png"
        with open(output_file, "wb") as f:
            f.write(png_bytes)
        # print(f"ðŸ’¾ Diagram saved as: {output_file}")

    except Exception:
        raise
    finally:
        if os.path.exists(CACHE_FOLDER):
            shutil.rmtree(CACHE_FOLDER)
            os.makedirs(CACHE_FOLDER)
    # with open(save_path, "wb") as f:
    #     f.write(png_bytes)
    # print(f"Saved diagram to {save_path}")

# ==========================
# Example usage
# ==========================
if __name__ == "__main__":
    user_prompt = "A user uploads data into Amazon S3. The data is then processed using AWS Glue, which stores curated results back into d-S3. From Glue, data flows into aws-glue-elastic-views, which feeds into Amazon Athena. Next, data is used in SageMaker. SageMaker metrics are monitored in CloudWatch, and results visualized in QuickSight."
    # user_prompt=" a user uploads data into s3 and then automatically triger lambda function and then it call sagemaker pipeline and it train and build model and deploy it into sagemaker then its output will be saved into s3 then from s3 it goes to cloudwatch and from cloudwatch to glue"
    flow_string = natural_to_flow(user_prompt)
    print("Parsed flow:", flow_string)
    generate_architecture_diagram_from_prompt(flow_string, title="Generated Pipeline")
```
